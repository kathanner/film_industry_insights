## Create the tables
manually creates each table and column name/type or format/key; ensures accuracy

```python
import sqlite3
```

```python
#only run this once or it will throw an error if the table already exists

sqlite3.connect('movie.db')

conn = sqlite3.connect('movie.db')  
cur = conn.cursor()

# Create table - imdb_name_basics 
cur.execute('''CREATE TABLE imdb_name_basics
                 ([pid] INTEGER,
                 [nconst] TEXT PRIMARY KEY,
                 [primary_name] TEXT, 
                 [birth_year] INTEGER, 
                 [death_year] INTEGER, 
                 [primary_profession] TEXT, 
                 [known_for_titles] TEXT);
            ''') 

# Create table - imdb_title_akas 
cur.execute('''CREATE TABLE imdb_title_akas
                 ([pid] INTEGER PRIMARY KEY,
                 [title_id] TEXT, 
                 [ordering] INTEGER, 
                 [title] TEXT, 
                 [region] TEXT, 
                 [language] TEXT, 
                 [types] TEXT, 
                 [attributes] TEXT, 
                 [is_original_title] TEXT);
            ''') 

# Create table - imdb_title_basics 
cur.execute('''CREATE TABLE imdb_title_basics
                 ([pid] INTEGER,
                 [tconst] TEXT PRIMARY KEY, 
                 [primary_title] TEXT, 
                 [original_title] TEXT, 
                 [start_year] INTEGER, 
                 [runtime_minutes] REAL, 
                 [genres] TEXT);
            ''') 

# Create table - imdb_title_principals 
cur.execute('''CREATE TABLE imdb_title_principals
                 ([pid] INTEGER PRIMARY KEY,
                 [tconst] TEXT, 
                 [ordering] INTEGER, 
                 [nconst] TEXT, 
                 [category] TEXT, 
                 [job] TEXT, 
                 [characters] TEXT);
            ''') 

# Create table - imdb_title_crew 
cur.execute('''CREATE TABLE imdb_title_crew
                 ([pid] INTEGER,
                 [tconst] TEXT PRIMARY KEY, 
                 [directors] TEXT,
                 [writers] TEXT);
            ''') 

# Create table - bom_movie_gross 
cur.execute('''CREATE TABLE bom_movie_gross
                 ([pid] INTEGER PRIMARY KEY,
                 [title] TEXT, 
                 [studio] TEXT, 
                 [domestic_gross] REAL, 
                 [foreign_gross] NUMERIC, 
                 [year] INTEGER);
            ''') 

# Create table - tmdb_movies 
cur.execute('''CREATE TABLE tmdb_movies
                 ([pid] INTEGER PRIMARY KEY,
                 [Unnamed: 0] INTEGER,
                 [genre_ids] INTEGER, 
                 [id] INTEGER, 
                 [original_language] TEXT, 
                 [original_title] TEXT,
                 [popularity] REAL, 
                 [release_date] DATE, 
                 [title] TEXT, 
                 [vote_average] REAL, 
                 [vote_count] INTEGER);
            ''')

# Create table - tn_movie_budgets 
cur.execute('''CREATE TABLE tn_movie_budgets
                 ([pid] INTEGER PRIMARY KEY,
                 [id] TEXT, 
                 [release_date] TEXT, 
                 [movie] TEXT,
                 [production_budget] INTEGER, 
                 [domestic_gross] INTEGER,
                 [worldwide_gross] INTEGER);
            ''') 

# Create table - rt_reviews 
cur.execute('''CREATE TABLE rt_reviews
                 ([pid] INTEGER PRIMARY KEY,
                 [Unnamed: 0] INTEGER,
                 [id] INTEGER,
                 [review] TEXT,
                 [rating] TEXT, 
                 [fresh] TEXT, 
                 [critic] TEXT, 
                 [top_critic] INTEGER,
                 [publisher] TEXT, 
                 [date] TEXT);
            ''')

# Create table - rt_movie_info 
cur.execute('''CREATE TABLE rt_movie_info
                 ([pid] INTEGER PRIMARY KEY,
                 [Unnamed: 0] INTEGER, 
                 [id] INTEGER, 
                 [synopsis] TEXT,
                 [rating] TEXT,
                 [genre] TEXT,
                 [director] TEXT,
                 [writer] TEXT,
                 [theater_date] TEXT,
                 [dvd_date] TEXT, 
                 [currency] TEXT, 
                 [box_office] INTEGER, 
                 [runtime] TEXT,
                 [studio] TEXT);
            ''')


conn.commit()

# type options: TEXT, NUMERIC, INTEGER, REAL, BLOB
    # NUMERIC column, the storage class of the text is converted to INTEGER or REAL (in order of preference)

# PRIMARY KEY: SQLite wants this to be unique; throws error otherwise

# To create a PRIMARY KEY adding two columns when a table already exists:
# ALTER TABLE tablenameHERE 
    # ADD CONSTRAINT primaryKeyNameHERE PRIMARY KEY (column1NameHERE, column2NameHERE)

# To Delete a PRIMARY KEY: 
# ALTER TABLE tablenameHERE
    # DROP CONSTRAINT primaryKeyNameHERE;
```

## Insert the values
write the data from the zip files to each established table

```python
# Insert the values from the csv file into the tables
imdb_name_basics.to_sql('imdb_name_basics', conn, if_exists='append', index_label= "pid") 
imdb_title_akas.to_sql('imdb_title_akas', conn, if_exists='append', index_label= "pid")
imdb_title_basics.to_sql('imdb_title_basics', conn, if_exists='append', index_label= "pid")
imdb_title_principals.to_sql('imdb_title_principals', conn, if_exists='append', index_label= "pid")
imdb_title_crew.to_sql('imdb_title_crew', conn, if_exists='append', index_label= "pid")
bom_movie_gross.to_sql('bom_movie_gross', conn, if_exists='append', index_label= "pid")
tmdb_movies.to_sql('tmdb_movies', conn, if_exists='append', index_label= "pid")
tn_movie_budgets.to_sql('tn_movie_budgets', conn, if_exists='append', index_label= "pid")
rt_reviews.to_sql('rt_reviews', conn, if_exists='append', index_label= "pid")
rt_movie_info.to_sql('rt_movie_info', conn, if_exists='append', index_label= "pid")

# if_exists='replace' --> Replaces the values from the csv file into the table
# if_exists='append' --> Insert the values from the csv file into the table
```

```python
# split columns with substr as needed
# drop unnecessary columns:
    # execute only columns needed in the df. then rewrite/replace df
    
conn = sqlite3.connect('movie.db') 
cur = conn.cursor()

cur.execute(""" 
SELECT pid, tconst, directors,
  substr(directors, 1,          instr(directors, ',') -1) AS director1,
  substr(directors, 11,          instr(directors, ',') -1) AS director2,
  substr(directors, 21,          instr(directors, ',') -1) AS director3,
  substr(directors, 31,          instr(directors, ',') -1) AS director4,
  writers,
  substr(writers, 1,          instr(writers, ',') -1) AS writer1,
  substr(writers, 11,          instr(writers, ',') -1) AS writer2,
  substr(writers, 21,          instr(writers, ',') -1) AS writer3,
  substr(writers, 31,          instr(writers, ',') -1) AS writer4
FROM imdb_title_crew
ORDER BY directors DESC
;
""")

dft = pd.DataFrame(cur.fetchall())
dft.columns = (x[0] for x in cur.description)
#print(dft.shape)
#print(dft.info())
dft.to_sql('imdb_title_crew', conn, if_exists='replace', index = False)

dft.head(10)
```
